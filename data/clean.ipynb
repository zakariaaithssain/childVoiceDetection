```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv("cv-valid-train.csv")
df.head()
```

```python
cleaned = df[["filename", "age", "gender"]]
cleaned.dropna(subset=["age"], ignore_index=True, inplace=True)
cleaned["label"] = np.where(cleaned["age"] == "teens", "child", "adult")
cleaned.drop(columns=["age", "gender"], inplace=True)
cleaned.head()
```

```python
cleaned.to_csv("labeled-files.csv")
```
